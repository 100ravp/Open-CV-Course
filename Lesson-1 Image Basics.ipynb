```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

# How to create a image ?

```python
#using numpy to create image
image = np.zeros((512,512,3))
```

```python
#risizing the figure and plotting
plt.figure(figsize = (6,6))
plt.imshow(image)
```

```python
#creating a white image 
image = np.ones((512,512,3))

#plotting 
plt.figure(figsize = (6,6))
plt.imshow(image)
```

# Uploading an Image from the local Machine

```python
#uploading the image
image = cv2.imread("IMG_1605.HEIC")
```

```python
#checking the shape
image.shape
```

```python
#pixel values
image
```

```python
#plotting the image
plt.figure(figsize = (6,6))
plt.imshow(image)
```

```python
#transforming to actual color
transformed = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (6,6))
plt.imshow(transformed)
```

```python
#converting to grayscale
gray_scale = cv2.imread("IMG_1605.HEIC",cv2.IMREAD_GRAYSCALE)
```

```python
#plotting
plt.figure(figsize = (6,6))
plt.imshow(gray_scale,cmap = "gray")
```

# Basic Transformations on image

```python
#resizing
resized = cv2.resize(transformed,(2000,2000))

plt.figure(figsize = (6,6))
plt.imshow(resized)
```

```python
#flipping
flipped_image = cv2.flip(transformed,1)
plt.imshow(flipped_image)
```

```python
#saving the image
cv2.imwrite("new.jpg",flipped_image)
```

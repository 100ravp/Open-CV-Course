```python
#libraries 
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#plotting function
def show_image(image):
    
    plt.figure(figsize = (8,8))
    plt.imshow(image)
```

```python
#creating image
image = np.zeros((512,512,3))
```

```python
show_image(image)
```

```python
#drawing Rectangle
cv2.rectangle(image,pt1 = (200,200),pt2 =(312,312),color = (0,255,0),thickness=5)
show_image(image)
```

```python
#drawing rectangle
cv2.rectangle(image,pt1 = (100,100),pt2 =(400,400),color = (0,255,0),thickness=3)
show_image(image)
```

```python
#drawing circle
cv2.circle(image,center = (100,100),radius=50, color=(255,0,0), thickness=5)
show_image(image)
```

```python
#drawing circle
cv2.circle(image,center = (100,400),radius=50, color=(255,0,0), thickness=5)
show_image(image)
```

```python
#loading meme template
image = cv2.imread("Drake-Hotline-Bling.jpg")
```

```python
show_image(image)
```

```python
#creating meme
corrected = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

font = cv2.FONT_ITALIC

cv2.putText(corrected, text ="Using Python for ML/AI", org =(600,200), fontFace =font, fontScale = 1.5 , color =(0,0,0),
            thickness = 3, lineType = cv2.LINE_AA)
cv2.putText(corrected, text ="Using Python for Memes", org =(600,800), fontFace =font, fontScale = 1.5, color =(0,0,0),
            thickness = 3, lineType = cv2.LINE_AA)

show_image(corrected) 
```

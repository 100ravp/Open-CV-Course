```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def plot_image(image):
    plt.figure(figsize = (7,7))
    plt.imshow(image)
```

```python
image = cv2.imread("Albert-Einstein.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plot_image(image)
```

# Custom made filter

```python
kernel = np.ones((20,20))/400
```

```python
new = cv2.filter2D(image, -1, kernel)

plot_image(new)
```

# Average Blur

```python
image = cv2.imread("Albert-Einstein.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
blur = cv2.blur(image, ksize = (20,20))
plot_image(blur)
```

# Gaussian Blur

```python
image = cv2.imread("Albert-Einstein.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
blur = cv2.GaussianBlur(image, (7,7),15)
plot_image(blur)
```

# Median Blur

```python
image = cv2.imread("Albert-Einstein.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
blur =  cv2.medianBlur(image,19)
plot_image(blur)
```

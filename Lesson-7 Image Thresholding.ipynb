```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def image_show(image):
    plt.figure(figsize = (8,8))
    plt.imshow(image,"gray")
```

```python
image = cv2.imread("58c074d8a1ca3631340ec41f2024efef.jpeg")
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_show(image)
```

```python
image = cv2.imread("58c074d8a1ca3631340ec41f2024efef.jpeg",0)
image_show(image)
```

```python
ret,thres = cv2.threshold(image,90,255,cv2.THRESH_BINARY)
image_show(thres)
```

```python
ret,thres = cv2.threshold(image,100,255,cv2.THRESH_BINARY_INV)
image_show(thres)
```

```python
ret,thres = cv2.threshold(image,100,255,cv2.THRESH_TRUNC)
image_show(thres)
```

```python
ret,thres = cv2.threshold(image,100,255,cv2.THRESH_TOZERO)
image_show(thres)
```

```python
ret,thres = cv2.threshold(image,90,255,cv2.THRESH_TOZERO_INV)
image_show(thres)
```

```python
thres = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,21,5)
image_show(thres)
```

```python
thres = cv2.adaptiveThreshold(image,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,21,5)
image_show(thres)
```

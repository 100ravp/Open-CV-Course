```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def image_plot(image):
    plt.figure(figsize = (10,10))
    plt.imshow(image, "gray")
```

```python
image = np.zeros((700,700))
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(image, text = "SAURAV", org = (100,350), fontFace= font, fontScale=4, color = (255,255,2550),
           thickness = 22, lineType = cv2.LINE_AA)

image_plot(image)
```

```python
kernel = np.ones((5,5))

kernel
```

```python
erode = cv2.erode(image, kernel, iterations =3)

image_plot(erode)
```

```python
dilation = cv2.dilate(image, kernel, iterations = 3)

image_plot(dilation)
```

```python
noise_image = cv2.imread("Noisy_background.png",0)

image_plot(noise_image)
```

```python
opening = cv2.morphologyEx(noise_image, cv2.MORPH_OPEN, kernel)

image_plot(opening)
```

```python
noise_image = cv2.imread("Noisy_foreground.png",0)

image_plot(noise_image)
```

```python
closing = cv2.morphologyEx(noise_image, cv2.MORPH_CLOSE, kernel)

image_plot(closing)
```

```python
gradient = cv2.morphologyEx(image, cv2.MORPH_GRADIENT, kernel)

image_plot(gradient)
```

```python
new_image = dilation- erode
```

```python
image_plot(new_image)
```
